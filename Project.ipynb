```python
a1 = sns.countplot(x = "Gender",data = df)
plt.title("Gender Count")
plt.xlabel("")
plt.ylabel("")
for i in a1.containers:
    a1.bar_label(i)
plt.show()
```

```python
plt.figure(figsize = (10,6))
a2 = sns.countplot(x = "Age Group",data = df,hue = "Gender")
plt.title("Age Group Wise Gender Count")
plt.xlabel("")
plt.ylabel("")
for i in a2.containers:
    a2.bar_label(i)
plt.show()
```

```python
Zone_Data = df["Zone"].value_counts()
#print(Zone_Data)

plt.pie(Zone_Data, labels=Zone_Data.index, autopct="%1.1f%%",startangle = 280)
plt.title("Zone Distribution")
plt.show()
```

```python
Human_age = df["Gender"].value_counts()
Human_age
plt.pie(Human_age,labels = Human_age.index,autopct = "%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```
